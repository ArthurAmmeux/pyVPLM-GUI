```python
# Import pyVPLM modules/functions
from pyvplm.core.definition import PositiveParameter, PositiveParameterSet
from pyvplm.addon.variablepowerlaw import buckingham_theorem

# Import libs
import numpy
import ipywidgets as widgets
import ipysheet

i = 0
Clicked = False

def add_variable(b):
    global i
    global Clicked
    if Clicked == True:
        sheet1.rows += 1
    else:
        Clicked = True
    ipysheet.row(i, [v_name.value, v_symbol.value, v_unit.value, v_lower_bound.value, v_upper_bound.value, Switch("Regular")])
    i += 1
    
def rem_variable(b):
    global i
    global Clicked
    if Clicked:
        if i > 1:
            sheet1.rows -= 1
            sheet1.cells = sheet1.cells[:-1]
            i -= 1
        elif i == 1:
            ipysheet.row(0,["", "", "", 0, 0, Switch("Output")])
            i = 0
            Clicked = False
        
def read_sheet(b):
    array = ipysheet.to_array(sheet1)
    print(array)
    
class State:
    def __init__(self, state_name):
        self.description = "Regular"
        self.tooltip = "Regular variable"
        self.color = "green"
        if state_name == "Output":
            self.description = "Output"
            self.tooltip = "Chosen output parameter"
            self.color = "orange"
        elif state_name == "Priority":
            self.description = "Priority"
            self.tooltip = "Priority variable"
            self.color = "blue"
    

class Switch(widgets.Button):
    def __init__(self, state_name):
        self.state = State(state_name)
        super().__init__(description=self.state.description, color=self.state.color, tooltip=self.state.tooltip)
        
    def set_state(self, state_name):
        self.state = State(state_name)
        self.description = self.state.description
        self.color = self.state.color
        self.tooltip = self.state.tooltip
        
    
v_name = widgets.Text(placeholder='length', description='Name:')
v_lower_bound = widgets.FloatText(placeholder='0', description='Lower bound:', layout=widgets.Layout(width='150px'))
v_upper_bound = widgets.FloatText(placeholder='100', description='Upper bound:', layout=widgets.Layout(width='150px'))
v_symbol = widgets.Text(placeholder='x', description='Symbol:', layout=widgets.Layout(width='150px'))
v_unit = widgets.Text(placeholder='m', description='Unit:', layout=widgets.Layout(width='150px'))

button1 = widgets.Button(
    description='Save variable',
    disabled=False,
    button_style='success', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Save variable')
button1.on_click(add_variable)

button2 = widgets.Button(
    description='Remove last',
    disabled=False,
    button_style='danger', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Remove last variable')
button2.on_click(rem_variable)

button3 = widgets.Button(description='To array',
    disabled=False,
    button_style='info', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='To array')
button3.on_click(read_sheet)

sheet1 = ipysheet.sheet(rows = 1, columns = 6, column_headers = ["Description", "Name", "Unit", "Lower bound", "Upper bound","Type/Priority"])
ipysheet.row(0,["", "", "", 0, 0, Switch("Output")])

up_box = widgets.HBox([v_name, v_symbol, v_unit])
down_box = widgets.HBox([v_lower_bound, v_upper_bound, button1, button2])
vbox_1 = widgets.VBox([up_box, down_box, sheet1, button3])
vbox_1
```
